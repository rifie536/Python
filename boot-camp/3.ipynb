```python
import numpy as np
import pandas as pd

df = pd.read_csv("./data/adult_income_dataset.csv")
```

```python
df["occupation"].unique()
```

```python
new_df = df.copy()
new_df["education"] = df["education"].replace("Bachelors", "Bachelor degree")
print(df["education"].unique())
print(new_df["education"].unique())
```

```python
new_df = df.copy()
new_df["net-capital-gain"] = new_df["capital-gain"] - new_df["capital-loss"]

new_df[["capital-gain", "capital-loss", "net-capital-gain"]].head()
```

```python
from sklearn.preprocessing import StandardScaler

new_df = df.copy()

scaler = StandardScaler()
data_standardized = scaler.fit_transform(new_df[["age"]])
new_df["age_standardized"] = data_standardized
new_df[["age", "age_standardized"]].head()
```

```python
new_df = df.copy()
Q1 = new_df["hours-per-week"].quantile(0.25)
Q3 = new_df["hours-per-week"].quantile(0.75)

IQR = Q3 - Q1

outlier_count = (
    (new_df["hours-per-week"] < (Q1 - 1.5 * IQR)) |
    (new_df["hours-per-week"] > (Q3 + 1.5 * IQR))
    ).sum()

print("hours-per-weekに含まれる外れ値の数:", outlier_count)

new_df["hours-per-week"] = np.where(
    new_df["hours-per-week"] < (Q1 - 1.5 * IQR), Q1, new_df["hours-per-week"]
)

new_df["hours-per-week"] = np.where(
    new_df["hours-per-week"] > (Q1 + 1.5 * IQR), Q3, new_df["hours-per-week"]
)

new_df[["hours-per-week"]].head()
```

```python
new_df = df.copy()

print("補完前:", new_df["occupation"].isnull().sum())
new_df["occupation"] = new_df["occupation"].fillna("None")
print("補完後:", new_df["occupation"].isnull().sum())
```

```python
new_df = df.copy()
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
new_df["hours-per-week-normalized"] = scaler.fit_transform(new_df[["hours-per-week"]])
new_df[["hours-per-week", "hours-per-week-normalized"]].head()
```
