```python
def calc_avg(nums):
    total = sum(nums)
    count = len(nums)
    return total / count

list = [10, 20, 30, 40, 50]
avg = calc_avg(list)
print("平均値:", avg)
```

```python
calc_avg = lambda nums: sum(nums) / len(nums)

list = [10, 20, 30, 40, 50]
avg = calc_avg(list)
print("平均値:", avg)
```
